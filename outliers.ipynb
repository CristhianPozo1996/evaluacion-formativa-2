**5)** Se va a generar una matriz de (N° Filas, N° Columnas) con números aleatorios (previamente fijos con seed(42)).

- Si se usa *np.random.rand* va a solicitar números aleatorios entre 0 y 1

- Si se usa *np.random.randn* va a solicitar números aleatorios que siguen una distribución normal (Campana de Gauss)

```python
#5
np.random.randn(100, 2)
```

**6)** Se va a generar una matriz llamada *X_inliers* multiplicada por 0.3 a partir de la matriz generada con np.random.randn(#Fila, #Col).

Esta nueva matriz, tiene las mismas dimesiones que la matriz de origen.

```python
#6
X_inliers = 0.3 * np.random.randn(100, 2)

X_inliers
```

**7)** Se va a modificar la matriz *X_inliers*.
- Se van a tomar los valores originales de la matriz, y se los va a sumar 2 (+2).
- Se van a tomar los valores originales de la metriz, y se los va a restar 2 (-2).

- Los valores sumados y restados se los va a unir en la matriz, teniendo el doble de datos.

```python
#7
X_inliers = np.r_[X_inliers + 2, X_inliers - 2]

X_inliers
```

**8)** Se va a generar una matriz de (N° Filas, N° Columnas) con números aleatorios (previamente fijos con seed(42)).

- Si se usa *np.random.rand* va a solicitar números aleatorios entre 0 y 1

- Si se usa *np.random.randn* va a solicitar números aleatorios que siguen una distribución normal (Campana de Gauss)

- Si se usa *np.random.uniform* va a solicitar números aleatorios entre un valor mínimo (low) y máximo (high)

```python
#8
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))

X_outliers
```

**9)** Se va a unir la matriz *X_inliers* y bajo esta la *X_outliers*

- *X_inliers* con *X_outliers* se van a almacenar en la variable *X*

```python
#9
X = np.r_[X_inliers, X_outliers]

X
```

**10)** Se va a contabilizar la cantidad de puntos (Filas) de la variable *X*, ese valor lo guardara en la variable *n*

```python
#10
n = len(X)

n
```

**11)** Se va a contabilizar la cantidad de puntos (Filas) de la variable *X_inliers*, ese valor lo guardara en la variable *n_inliers*

```python
#11
n_inliers = len(X_inliers)

n_inliers
```

**12)** Se va a contabilizar la cantidad de puntos (Filas) de la variable *X_outliers*, ese valor lo guardara en la variable *n_outliers*

```python
#12
n_outliers = len(X_outliers)

n_outliers
```

**13)** Se va a crear un array *ground_truth* con la cantidad de puntos que haya en la matriz *x*

-Usando el submodulo *np.ones* de la librería *numpy (np)*, va a llenar los datos con el número 1.

NOTA: Estamos haciendo que todos nuestros datos sean 1.

```python
#13
ground_truth = np.ones(len(X), dtype=int)

ground_truth
```

**14)** Se va a modificar el array *ground_truth* colocando a la CANTIDAD de la variable *n_outliers* un valor de *-1*

```python
#14
ground_truth[-n_outliers:] = -1

ground_truth
```

**15)** Se va a utilizar la clase (algoritmo) *LocalOutlierFactor* del submodulo *neighbors* de la libería *sklearn* y se lo va a almacenar en la variable *clf*

- n_neighbors=#Número: Para cada punto evalua sus #Número vecinos más cercanos
- contamination=#Número: Indica que porcentaje estimado es anómalo (si 0.1 se espera que el 10% de datos son outliers)

```python
#15
clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1)

clf
```

**16)** Entrena al modelo LOF y predice si cada punto es inlier o outlier.

- *fit* permite que LocalOutlierFactor aprenda la estructura de los datos.
- *predict* Después de entrenarse, devueve una predicción y la guarda en la variable *y_pred*

```python
#16
y_pred = clf.fit_predict(X)

y_pred
```

**17)** Compara la predicción hecha vs la realidad
- Predicción: y_pred
- Realidad: ground_truth

Con .sum() suma todos los valores TRUE que es igual al número de errores del MODELO LOF.

```python
#17
n_errors = (y_pred != ground_truth).sum()

n_errors
```
